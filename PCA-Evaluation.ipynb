## **This Notebook generates performance graph for n Principle Components for our selected models from pca_acc.txt file**

### **Importing Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### **Read data from pca_acc.txt file**

```python
with open('pca_acc.txt', 'r+') as f:
    lines = f.readlines()
```

### **Pre-Process Data for graph**

```python
data = []
for line in lines:
    ele = line.split(' ')
    data.append(ele)
    
df = pd.DataFrame(data, columns=['Components', 'LR', 'RF', 'XGB', 'SVM'])              
df['SVM'] = df['SVM'].str.replace('\n','')
df = df.astype(float)
df = df.iloc[:40, :] ##
df.head()
```

### **Plotting the Graph**

```python
fig, ax = plt.subplots(figsize=(16,10))

ax.plot(df['Components'], df['LR'], color = 'red', label = 'Linear Regression')
ax.plot(df['Components'], df['RF'], color = 'blue', label = 'Random Forest')
ax.plot(df['Components'], df['XGB'], color = 'green', label = 'XGBoost')
ax.plot(df['Components'], df['SVM'], color = 'brown', label = 'SVM')

ax.legend()

plt.ylim(0, 1)
plt.title('Number of components vs Accuracy', fontsize = 20, fontweight = 'bold', fontname = 'Times New Roman')
plt.xlabel('No. of components', fontsize = 15, fontweight = 'bold', fontname = 'Times New Roman')
plt.ylabel('Accuracy', fontsize = 15, fontweight = 'bold', fontname = 'Times New Roman')
plt.show()
```
